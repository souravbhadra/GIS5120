```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def generate_csr(sample_size, area=(0, 1, 0, 1)):
    
    xmin, xmax, ymin, ymax = area
    
    # Generate random points within the area
    x_coords = np.random.uniform(xmin, xmax, sample_size)
    y_coords = np.random.uniform(ymin, ymax, sample_size)
    
    return x_coords, y_coords
```

```python
sample_size = 1000  # Adjust this value for different sample sizes
area = (0, 10, 0, 10)  # Define a custom area, here 10x10 units
```

```python
# Generate the CSR points
x_coords, y_coords = generate_csr(sample_size, area)
```

```python
font = {'family' : 'Arial',
        'size'   : 7}
plt.rc('font', **font)
```

```python
fig, ax = plt.subplots(1, 3, figsize=(15,4.5), sharey=True, dpi=300)

sample_sizes = [100, 1000, 10000]

for i, sample_size in enumerate(sample_sizes):
    # Generate the CSR points
    x_coords, y_coords = generate_csr(sample_size, area)
    ax[i].scatter(
        x_coords, y_coords, 
        facecolors='none', 
        edgecolors='black', 
        s=5
    )
    ax[i].set_title(f'Sample Size = {sample_size}', fontsize=12)
    ax[i].set_xticks([])
    ax[i].set_yticks([])

plt.subplots_adjust(wspace=0.05)
plt.show()
```

```python
fig, ax = plt.subplots(1, 1, figsize=(3, 3), dpi=300)

x_coords, y_coords = generate_csr(100, area)
ax.scatter(
    x_coords, y_coords, 
    facecolors='none', 
    edgecolors='black', 
    s=5
)
ax.set_xticks([])
ax.set_yticks([])
ax.set_xlim(0, 30)
ax.set_ylim(0, 30)

plt.show()
```
