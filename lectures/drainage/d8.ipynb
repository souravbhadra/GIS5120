```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap, BoundaryNorm
```

```python
font = {'family' : 'Arial',
        'size'   : 16}
plt.rc('font', **font)
```

```python
elev = np.array([
    [78, 72, 69, 71, 58, 49],
    [74, 67, 56, 49, 46, 50],
    [69, 53, 44, 37, 38, 48],
    [64, 58, 55, 22, 31, 24],
    [68, 61, 47, 21, 16, 19],
    [74, 53, 34, 12, 11, 12]
])
```

```python
fig, ax = plt.subplots(1, 1, dpi=300, figsize=(5,5))
sns.heatmap(
    elev, cmap='rainbow', annot=True,
    linewidths=0.5, linecolor='black',
    ax=ax, cbar=False
)
ax.axis('off')
plt.show()
```

```python
encoder = np.array([
    [32, 64, 128, np.nan, np.nan, np.nan],
    [16, np.nan, 1, np.nan, np.nan, np.nan],
    [8, 4, 2, np.nan, np.nan, np.nan],
    [np.nan, np.nan, np.nan, np.nan, np.nan, np.nan],
    [np.nan, np.nan, np.nan, np.nan, np.nan, np.nan],
    [np.nan, np.nan, np.nan, np.nan, np.nan, np.nan]
])
```

```python
value_to_color = {
    1: 'limegreen',
    2: 'darkgreen',
    4: 'skyblue',
    8: 'deepskyblue',
    16: 'dodgerblue',
    32: 'violet',
    64: 'red',
    128: 'darkorange',
    np.nan: '#ffffff'  
}
```

```python
# Step 3: Create a list of colors based on the unique values in your data
unique_values = [1, 2, 4, 8, 16, 32, 64, 128]
colors = [value_to_color[val] for val in unique_values]
cmap = ListedColormap(colors)

# Step 4: Define the boundaries for each value
bounds = [0.5, 1.5, 2.5, 4.5, 8.5, 16.5, 32.5, 64.5, 128.5]
norm = BoundaryNorm(bounds, cmap.N)
```

```python
fig, ax = plt.subplots(1, 1, dpi=300, figsize=(5,5))
sns.heatmap(
    encoder, cmap=cmap, norm=norm, annot=True,
    linewidths=0.5, linecolor='black', fmt='.0f',
    ax=ax, cbar=False
)
ax.axis('off')
plt.show()
```
