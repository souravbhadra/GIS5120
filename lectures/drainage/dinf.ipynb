```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap, BoundaryNorm
```

```python
font = {'family' : 'Arial',
        'size'   : 16}
plt.rc('font', **font)
```

```python
elev = np.array([
    [50, 39, 35, np.nan, np.nan, np.nan],
    [52, 47, 45, np.nan, np.nan, np.nan],
    [55, 49, 48, np.nan, np.nan, np.nan],
    [np.nan, np.nan, np.nan, np.nan, np.nan, np.nan],
    [np.nan, np.nan, np.nan, np.nan, np.nan, np.nan],
    [np.nan, np.nan, np.nan, np.nan, np.nan, np.nan]
])
```

```python
fig, ax = plt.subplots(1, 1, dpi=300, figsize=(5,5))
sns.heatmap(
    elev, cmap='rainbow', annot=True,
    linewidths=0.5, linecolor='black',
    ax=ax, cbar=False
)
ax.axis('off')
plt.show()
```
